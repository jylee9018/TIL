```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Graph Title", loc="center", pad=50)

plt.show()
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], label="Sqare")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.legend()

plt.show()
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], label="Sqare")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.xlim([0, 30])
plt.ylim([0, 30])
plt.legend()

plt.show()
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], label="Sqare")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 30, 0, 30])
plt.legend()

plt.show()
```

```python
plt.plot([1, 2, 3], [1, 1, 1], "-", color="C0", label="Solid")
plt.plot([1, 2, 3], [2, 2, 2], "--", color="C1", label="Dashed")
plt.plot([1, 2, 3], [3, 3, 3], "-.", color="C2", label="Dash-dot")
plt.plot([1, 2, 3], [4, 4, 4], ":", color="C3", label="Dotted")

plt.plot([1, 2, 3], [6, 6, 6], linestyle="solid", color="C0", label="Dotted")
plt.plot([1, 2, 3], [7, 7, 7], linestyle="dashed", color="C1", label="Dashed")
plt.plot([1, 2, 3], [8, 8, 8], linestyle="dashdot", color="C2", label="Dash-dot")
plt.plot([1, 2, 3], [9, 9, 9], linestyle="dotted", color="C3", label="Dotted")

plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 4, 0, 12])
plt.legend(loc="upper right", ncol=4)

plt.show()
```

```python
# 튜플을 사용하여 선 종류 커스터마이즈
# (offset, (on_off_seq))

plt.plot([1, 2, 3], [1, 1, 1], linestyle=(0, (1, 1)), color="C0", label="(0, (1, 1))")
plt.plot([1, 2, 3], [2, 2, 2], linestyle=(0, (1, 5)), color="C1", label="(0, (1, 5))")
plt.plot([1, 2, 3], [3, 3, 3], linestyle=(0, (5, 1)), color="C2", label="(0, (5, 1))")
plt.plot(
    [1, 2, 3],
    [4, 4, 4],
    linestyle=(0, (1, 5, 3, 5)),
    color="C3",
    label="(0, (1, 5, 3, 5))",
)

plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 4, 0, 5])
plt.legend(loc="upper right", ncol=2)
plt.tight_layout()

plt.show()
```

```python
plt.plot([0, 1, 2], label="No")
plt.plot([1, 2, 3], "o", label="o")  # Circle
plt.plot([2, 3, 4], "ko", label="ko")  # Black, Circle
plt.plot([3, 4, 5], "ko-", label="ko-")  # Black, Circle, Solid
plt.plot([4, 5, 6], "ko--", label="ko--")  # Black, Circle, Dotted

plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.legend(loc="upper left", ncol=3)
plt.tight_layout()

plt.show()
```

```python
plt.plot([1, 2, 3], label="No")
plt.plot([2, 3, 4], marker=1, label="marker=1")
plt.plot([3, 4, 5], marker="1", label="marker='1'")
plt.plot([4, 5, 6], marker="D", label="marker='D'")
plt.plot([5, 6, 7], marker="d", label="marker='d'")
plt.plot([6, 7, 8], marker="$A$", label="marker=$A$")

plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.legend(loc="upper left", ncol=3)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 20)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, ".-")
plt.title("1st Graph")

plt.subplot(1, 2, 2)
plt.plot(x, ".--")
plt.title("2nd Graph")

plt.show()
```
